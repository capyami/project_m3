```python
import pandas as pd
import numpy as np
```

```python
diamonds_labeling = pd.read_csv('../../../001data/003preprocessing/002feature_engineering/diamonds_fe.csv')
```

```python
diamonds_test_labeling = pd.read_csv('../../../001data/003preprocessing/002feature_engineering/diamonds_test_fe.csv')
```

```python
cat_cols = ['cut', 'color', 'clarity', 'city']
diamonds_labeling = diamonds_labeling[cat_cols]
diamonds_test_labeling = diamonds_test_labeling[cat_cols]
```

Labeling for cut, color and clarity

cut: (Fair, Good, Very Good, Premium, Ideal)
color: J (worst) to D (best)
clarity: (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
not to city, get_dummies

```python
encoding_cut = {'Fair':1, 'Good':3, 'Very Good':6, 'Premium':10, 'Ideal':15}
def ordinal_encoding_cut(x):
    for key in encoding_cut:
        if x == key:
            return encoding_cut[key]
```

```python
encoding_color = {'J':1.43, 'I':2.86, 'H':4.29, 'G':5.71, 'F':7.14, 'E':8.57, 'D':10}
def ordinal_encoding_color(x):
    for key in encoding_color:
        if x == key:
            return encoding_color[key]
```

```python
encoding_clarity = {'I1':1.25, 'SI2':2.50, 'SI1':3.75, 'VS2':5.00, 'VS1':6.25, 'VVS2':7.50, 'VVS1':8.75, 'IF':10}
def ordinal_encoding_clarity(x):
    for key in encoding_clarity:
        if x == key:
            return encoding_clarity[key]
```

```python
diamonds_labeling['cut'] = diamonds_labeling['cut'].apply(ordinal_encoding_cut)
diamonds_labeling
```

```python
diamonds_labeling['color'] = diamonds_labeling['color'].apply(ordinal_encoding_color)
diamonds_labeling
```

```python
diamonds_labeling['clarity'] = diamonds_labeling['clarity'].apply(ordinal_encoding_clarity)
diamonds_labeling
```

```python
diamonds_test_labeling['cut'] = diamonds_test_labeling['cut'].apply(ordinal_encoding_cut)
diamonds_test_labeling
```

```python
diamonds_test_labeling['color'] = diamonds_test_labeling['color'].apply(ordinal_encoding_color)
diamonds_test_labeling
```

```python
diamonds_test_labeling['clarity'] = diamonds_test_labeling['clarity'].apply(ordinal_encoding_clarity)
diamonds_test_labeling
```

get dummies for city

```python
diamonds_labeling = pd.get_dummies(diamonds_labeling, 
                                          columns=['city'], 
                                          drop_first=False)
diamonds_labeling.info()
```
