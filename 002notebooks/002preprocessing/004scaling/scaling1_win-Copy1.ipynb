```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import PowerTransformer
```

```python
diamonds_scaling_win = pd.read_csv('../../../../data/preprocessing/001feature_engineering/diamonds_fe.csv')
```

```python
diamonds_test_scaling_win = pd.read_csv('../../../../data/preprocessing/001feature_engineering/diamonds_test_fe.csv')
```

```python
diamonds_scaling_win['scaling'] = 1
diamonds_scaling_win = diamonds_scaling_win[['carat', 'depth', 'table', 'x', 'y',
       'z', 'xyz/depth', 'xyz/carat', 'depth/table', 'depth/z',
       'depth/carat', 'table/z', 'table/carat', 'carat/z', 'x/z', 'y/z',
       'x/y', 'density', 'volume', 'scaling']]
diamonds_scaling_win.info()

diamonds_scaling_win
```

```python
diamonds_test_scaling_win['scaling'] = 2


diamonds_test_scaling_win = diamonds_test_scaling_win[['carat', 'depth', 'table', 'x', 'y',
       'z', 'xyz/depth', 'xyz/carat', 'depth/table', 'depth/z',
       'depth/carat', 'table/z', 'table/carat', 'carat/z', 'x/z', 'y/z',
       'x/y', 'density', 'volume', 'scaling']]
diamonds_test_scaling_win
```

```python
#scaler = RobustScaler()
#scaler = MinMaxScaler()
#scaler = StandardScaler()
scaler = PowerTransformer()
```

```python
# concat
diamonds_scaling_win = pd.concat([diamonds_scaling_win, diamonds_test_scaling_win])
diamonds_scaling_win
```

```python
scaled_data = scaler.fit_transform(diamonds_scaling_win)
scaled_data
scaled_df = pd.DataFrame(scaled_data)
scaled_df.set_axis(['carat', 'depth', 'table', 'x', 'y',
       'z', 'xyz/depth', 'xyz/carat', 'depth/table', 'depth/z',
       'depth/carat', 'table/z', 'table/carat', 'carat/z', 'x/z', 'y/z',
       'x/y', 'density', 'volume', 'scaling'], axis='columns', inplace=True)
```

```python
# select data of train, delete test
scaled_df = scaled_df.loc[scaled_df['scaling']<0]
scaled_df
```

```python
scaled_data_test = scaler.fit_transform(diamonds_test_scaling_win)
scaled_data_test
```

```python
scaled_df_test = pd.DataFrame(scaled_data_test)
scaled_df_test.set_axis(['carat', 'depth', 'table', 'x', 'y',
       'z', 'xyz/depth', 'xyz/carat', 'depth/table', 'depth/z',
       'depth/carat', 'table/z', 'table/carat', 'carat/z', 'x/z', 'y/z',
       'x/y', 'density', 'volume', 'scaling'], axis='columns', inplace=True)
```
