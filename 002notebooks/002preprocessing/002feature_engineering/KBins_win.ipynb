## KBins with test

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
from sklearn.preprocessing import KBinsDiscretizer
```

```python
diamonds_kbin = pd.read_csv('../../../001data/003preprocessing/002feature_engineering/diamonds_fe.csv')
diamonds_kbin
```

```python
diamonds_kbin = diamonds_kbin[['proportion', 'depth/carat', 'table/carat', 'x/y', 'depth/table', 'volume/table', 'volume/depth']]
diamonds_kbin
```

```python
diamonds_test_kbin = pd.read_csv('../../../001data/003preprocessing/002feature_engineering/diamonds_test_fe.csv')
diamonds_test_kbin
```

```python
diamonds_test_kbin = diamonds_test_kbin[['proportion', 'depth/carat', 'table/carat', 'x/y', 'depth/table', 'volume/table', 'volume/depth']]
diamonds_test_kbin
```

```python
# concat with test
diamonds_kbin_w = pd.concat([diamonds_kbin, diamonds_test_kbin])
diamonds_kbin_w
```

```python
kbin_est = KBinsDiscretizer(n_bins=15, encode='ordinal', strategy='kmeans')
kbin_est.fit(diamonds_kbin_w)
diamonds_kbin_w = kbin_est.transform(diamonds_kbin_w)
```

```python
diamonds_kbin_w = pd.DataFrame(diamonds_kbin_w)
diamonds_kbin_w
```

```python
diamonds_kbin_w.set_axis([['proportion_kbw', 'depth/carat_kbw', 'table/carat_kbw', 'x/y_kbw', 'depth/table_kbw', 'volume/table_kbw', 'volume/depth_kbw']], axis='columns', inplace=True)
```

```python
diamonds_kbin_w
```

```python
diamonds_kbin_w = diamonds_kbin_w[:40455]
diamonds_kbin_w
```

```python
diamonds_kbin_w.to_csv('../../../001data/003preprocessing/002feature_engineering/diamonds_kbinsw.csv')
```
