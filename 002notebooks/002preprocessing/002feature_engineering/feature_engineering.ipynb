```python
def price_div(price, div):
    if div == 0:
        return 0
    else:
        return price/div
    
def densidad(mass_diamond, volumen):
    density_diamond = mass_diamond/volumen
    return density_diamond
    
def volumen(mass_diamond):
    volume_diamond = mass_diamond/3.51
    return volume_diamond
    
def masss(carat):
    mass_diamond = carat*0.20
    return mass_diamond

def multiply_divide(uno, dos, tres, cuatro):
    resultado = uno * dos * tres / cuatro
    return resultado

def proportion(x, y, z):
    mean_xy = x*y/2
    propor = mean_xy/z
    return propor
```

```python
def columns_diamonds(diamondsclean):
    diamondsclean['xyz/depth'] = diamondsclean.apply(lambda x : multiply_divide(x['x'], x['y'], x['z'], x['depth']), axis=1)
    diamondsclean['xyz/carat'] = diamondsclean.apply(lambda x : multiply_divide(x['x'], x['y'], x['z'], x['carat']), axis=1)
    diamondsclean['depth/table'] = diamondsclean.apply(lambda x : price_div(x['depth'], x['table']), axis=1)
    diamondsclean['depth/z'] = diamondsclean.apply(lambda x : price_div(x['depth'], x['z']), axis=1)
    diamondsclean['depth/carat'] = diamondsclean.apply(lambda x : price_div(x['depth'], x['carat']), axis=1)
    diamondsclean['table/z'] = diamondsclean.apply(lambda x : price_div(x['table'], x['z']), axis=1)
    diamondsclean['table/carat'] = diamondsclean.apply(lambda x : price_div(x['table'], x['carat']), axis=1)
    diamondsclean['carat/z'] = diamondsclean.apply(lambda x : price_div(x['carat'], x['z']), axis=1)
    diamondsclean['x/z'] = diamondsclean.apply(lambda x : price_div(x['x'], x['z']), axis=1)
    diamondsclean['y/z'] = diamondsclean.apply(lambda x : price_div(x['y'], x['z']), axis=1)
    diamondsclean['x/y'] = diamondsclean.apply(lambda x : price_div(x['x'], x['y']), axis=1)
    diamondsclean['mass'] = diamondsclean.apply(lambda x : masss(x['carat']), axis=1)
    diamondsclean['volume'] = diamondsclean.apply(lambda x : volumen(x['mass']), axis=1)
    diamondsclean['density'] = diamondsclean.apply(lambda x : densidad(x['mass'], x['volume']), axis=1)
    diamondsclean['proportion'] = diamondsclean.apply(lambda x : proportion(x['x'], x['y'], x['z']), axis =1)
    diamondsclean['carat/depth'] = diamondsclean.apply(lambda x : price_div(x['carat'], x['depth']), axis =1)
    diamondsclean['carat/table'] = diamondsclean.apply(lambda x : price_div(x['carat'], x['table']), axis =1)
    diamondsclean['volume/table'] = diamondsclean.apply(lambda x : price_div(x['volume'], x['table']), axis =1)
    diamondsclean['volume/depth'] = diamondsclean.apply(lambda x : price_div(x['volume'], x['depth']), axis =1)
    return diamondsclean
```

```python
diamonds = columns_diamonds(diamonds)
```
